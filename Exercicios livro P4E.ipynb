#### 8-
Unidade Central de processamento (CPU): É a parte do computador que pergunta "E agora?", executa instruções em linguagem de máquina, e é o equivalente ao cérebro.

Memória Principal: Armazena temporariamente informações que a CPU precisa com urgência, essa informação desaparece quando o computador é desligado. É equivalente a memória de curto prazo.

Memória Secundária: É uma memória capaz de armazenar informações mesmo depois do computador ser desligado (ex.: HD, USB, CDs e DVDs). Ela é muito mais lenta que a memória principal e é o equivalente a memória de longo prazo dos seres humanos.

Dispositivo de Entrada: São dispositivos conectados ao computador, como o teclado, microfones, mouse, touchpad etc. É o equivalente aos membrose órgãos sensoriais dos seres humanos.

Dispositivos de saída: São também conectados ao computador, são os monitores, caixas de som, etc. São o equivalente a boca e o olho.

# Capítulo 3: Execução Condicional

```python
def pagamento(horas = input("Digite as Horas:"), taxa = input("digite a taxa:")):
    h = float(horas)
    t = float(taxa)
    if h > 40:
        extra = (h - 40) * (1.5 * t)
        valor = (h - (h % 40)) * t + extra
        print("Pagamento:", valor)
    else:
        valor = h * t
        print("Pagamento:", valor)

pagamento()
```
